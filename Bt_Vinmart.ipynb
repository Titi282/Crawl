```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
```

```python
chrome_driver_path=r"C:\Users\Admin\Downloads\chromedriver_win32\chromedriver.exe"
driver = webdriver.Chrome(executable_path=chrome_driver_path)
url='https://vinmart.co/thit-46910/'
driver.get(url)
```

Lấy link sản phẩm của 10 trang đầu tiên

```python
all_meat_url=[]
for page in range(1,11):
    resp=requests.get(url+'?trang='+str(page))
    soup=BeautifulSoup(resp.content, "html.parser")
    meat_url=[('https://vinmart.co'+i['href']) for idx,i in enumerate(soup.select('div > a')) if (idx >=1 and idx<=20)]
    all_meat_url+=meat_url
all_meat_url
```

```python
with open('kq.txt','w') as f:
    for i in all_meat_url:
        f.writelines(f'{i[0]}, {i[1]}, {i[2]}')
```

```python
product_names=[] #Tên sản phẩm
stores=[] #Tten cửa hàng
prices=[] #Đơn giá
status=[] # Trạng thái
warehouse=[] # Kho hàng
describe=[] #Mô tả sản phẩm
for meat_url in all_meat_url:
    driver.get(meat_url)
    name=driver.find_element(By.XPATH,'/html/body/main/div/div[1]/div/div[2]/h1')
    store=driver.find_element(By.XPATH,'/html/body/main/div/div[1]/div/div[2]/div[1]/a[1]/strong')
    price=driver.find_element(By.XPATH,'//span[@class="special-price"]')
    st=driver.find_element(By.XPATH,'/html/body/main/div/div[1]/div/div[2]/div[2]')
    try:
        labe=driver.find_element(By.XPATH,'/html/body/main/div/div[1]/div/div[2]/div[3]/label')
        if labe.text=='Kho hàng':
            wh=driver.find_element(By.XPATH,'/html/body/main/div/div[1]/div/div[2]/div[3]/span')
            warehouse.append(wh.text)
    except:
        warehouse.append('Không có')
    dc=driver.find_element(By.XPATH,'//*[@id="des"]')
    product_names.append(name.text)
    stores.append(store.text)
    prices.append(price.text)
    status.append(st.text)
    describe.append(dc.text)
```

Xóa những ký tự đặc biệt, xóa khoảng trắng

```python
import re
describe_=[]
for i in describe:
    i=re.sub(r'[^\w]',' ',i)
    i=re.sub(r'\s+',' ',i)
    describe_.append(i)
```

```python
prices_=[]
for i in prices:
    prices_.append(re.sub('[^\d]','',i))
```

```python
status_=[]
for i in status:
    status_.append(re.sub('Trạng thái: ','',i))
```

```python
product_names_=[]
for i in product_names:
    i=re.sub(r'[^\w]',' ',i)
    i=re.sub(r'\s+',' ',i)
    product_names_.append(i)
```

```python
d={'Link':all_meat_url,'Tên sản phẩm':product_names_,'Cửa hàng':stores,'Giá bán':prices_,'Trạng thái':status_,'Kho hàng':warehouse,'Mô tả sản phẩm':describe_}
df=pd.DataFrame(d)
```
